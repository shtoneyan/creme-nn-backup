```python
for cell_line in ['K562', 'GM12878', 'PC-3']:
    contexts = pd.read_csv(f'../results/summary_csvs/enformer/{cell_line}_selected_contexts.csv')
    counter = {}
    for i, row in contexts.iterrows():
        if row['seq_id'].split('_')[0] in housekeeping_genes['Gene.name'].values:
            if row['context'] not in counter.keys():
                counter[row['context']] = 0
            counter[row['context']] += 1
    print(cell_line)
    N_df = contexts.groupby('context').count().reset_index()
    N_dict = dict(zip(N_df['context'], N_df['cell_line']))
    for k, v in counter.items():
        print(f"{k}: {np.round(v/N_dict[k]*100, 3)}% (N={v})")
```
